```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python

#we req the name, character, real name   in the dictionary

def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if(counter!=3):
      L.append(i['name'])
      counter+=1
    else :
        break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetchdirector(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(fetchdirector)
```

```python
#overview is a string,to  nake it into a list so tyhat we can concat with other list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove the spaces between the words bcoz if u have gudela hemanth,gudela neharika,instead of gudela heamnth movies u may get gudela nehaika mnoveis
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```
